```python
def user_inputs():
    s1=int(input("Enter the sub1:"))
    s2= int(input("Enter the sub2:"))
    s3 = int(input("Enter the sub3:"))
    email = input("Enter the email:")
    return s1,s2,s3,email

def avg_cal(x,y,z):
    total = x+y+z
    return total/3

def per_cal(a,b,c):
    per = (a+b+c)/300*100
    return per
def email_val(em):
    if em.endswith('@gmail.com'):
        return "valid"
    else:
        return "invalid"

def main():
    sub1,sub2,sub3,email = user_inputs()
    avg = avg_cal(sub1,sub2,sub3)
    
    per = per_cal(sub1,sub2,sub3)
    status = email_val(email)
    print("Subject marks:",sub1,sub2,sub3)
    print("Avg Marks:",avg)
    print("Per:",per)
    print("Email:",email,status)

if __name__=='__main__':
    main()
```
